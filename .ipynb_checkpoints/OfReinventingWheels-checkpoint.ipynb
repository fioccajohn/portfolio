## Comprehensize descriptive stats functions already exist.

Don't write your own extended pandas `describe` method. Use `statsmodels` instead.

```python
from statsmodels.stats.descriptivestats import describe
import pandas as pd
import numpy as np
from scipy.stats import norm, t

s = pd.Series(range(10))

measures = ["mean", "std_err", "nobs"]

desc_stats = s.pipe(describe).squeeze()[measures]
desc_stats.at["dof"] = desc_stats["nobs"] - 1

rv_norm = norm()
rv_t = t(desc_stats.loc["dof"])

# Upper and Lower CI


def compute_confidence_intervals(distribution, alpha = 0.95):
    lower_ci = (
        desc_stats["mean"] - distribution.ppf((1 - alpha) / 2) * desc_stats["std_err"]
    )
    upper_ci = (
        desc_stats["mean"] + distribution.ppf((1 - alpha) / 2) * desc_stats["std_err"]
    )
    return (lower_ci.item(), upper_ci.item())
```

## Alpha $\alpha$
$\alpha$ is a risk threshold: the probability of making an incorrect conclusion against what you believed before.
Technically stated, it represents the probability of rejecting a true null hypothesis (a Type 1 error).
In a confidence interval, it is the probability of not capturing the true value.

Power, or $\beta$, is the likelihood of detecting an effect that exists. Technically stated, is the probability of correctly rejecting a false null hypothesis.

```python
s.pipe(describe, alpha=0.10)
```

```python
s.pipe(describe, use_t=True)
```

```python
print(compute_confidence_intervals(rv_norm))
print(compute_confidence_intervals(rv_t))
```

`describe` defaults to a Z distribution. Pass `use_t=True` as a kwarg for small sample sizes.

```python
# Uses Z (Normal)
describe(s).filter(like="ci", axis=0).T
```

```python
# Uses T for small sample size.
describe(s, use_t=True).filter(like="ci", axis=0).T
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

%matplotlib inline

# Parameters
mu_null = 0  # Mean under the null hypothesis
mu_alt = 1.5  # Mean under the alternative hypothesis
std_dev = 1  # Standard deviation for both distributions
alpha = 0.05  # Significance level (Type I error rate)
z_critical = norm.ppf(1 - alpha)  # Z critical value for alpha
beta = 0.2  # Type II error rate, so power = 1 - beta
z_beta = norm.ppf(1 - beta)

# Range for plotting
x = np.linspace(-3, 5, 500)
null_dist = norm.pdf(x, mu_null, std_dev)
alt_dist = norm.pdf(x, mu_alt, std_dev)

# Calculate critical value for rejection region
critical_value = mu_null + z_critical * std_dev

# Plot
plt.figure(figsize=(12, 6))
plt.plot(x, null_dist, label="Null Distribution (H0)", color="blue")
plt.plot(x, alt_dist, label="Alternative Distribution (H1)", color="orange")

# Fill the rejection region (alpha) in the null distribution
plt.fill_between(x, 0, null_dist, where=(x >= critical_value), color="blue", alpha=0.3, label="Type I Error (α)")

# Fill the beta region (Type II error) in the alternative distribution
beta_region = mu_alt - z_beta * std_dev
plt.fill_between(x, 0, alt_dist, where=(x <= critical_value), color="orange", alpha=0.3, label="Type II Error (β)")

# Add labels and lines
plt.axvline(critical_value, color='black', linestyle='--', label="Critical Value")
plt.axvline(beta_region, color='purple', linestyle=':', label="Beta Threshold")
plt.title("Alpha (Type I Error) and Beta (Type II Error) in Hypothesis Testing")
plt.xlabel("Test Statistic Value")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(alpha=0.2)
```
