```python
from scipy import stats
import numpy as np
import seaborn as sns
```

```python
%matplotlib inline
```

```python
norm_rv = stats.norm()
```

```python
samples = norm_rv.rvs(100)

kinds = ['hist', 'kde', 'ecdf']

sns.boxplot(samples, orient='h')

for k in kinds:
    sns.displot(samples, kind=k, rug=True)



# sns.violinplot(samples)
```

```python
norm_rv.support()

x = 0
percentile = .5

[norm_rv.pdf(x),
norm_rv.cdf(x),
norm_rv.sf(x),
norm_rv.isf(percentile),
norm_rv.ppf(percentile),]
```
